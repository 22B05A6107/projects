```python
from diffusers import StableDiffusionPipeline
import torch

# Load the Stable Diffusion model
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)

# Check if CUDA is available before moving the model to GPU
if torch.cuda.is_available():
    pipe = pipe.to("cuda")  # Use GPU for faster performance
    print("Model loaded to GPU.")
else:
    print("CUDA not available, running on CPU.")
```

```python
# Generate an image based on a text prompt
prompt = "A futuristic cityscape with flying cars and glowing neon lights"
image = pipe(prompt).images[0]

# Save the image with a correct file path
image.save(r"C:\Users\Admin\Downloads\sirisha.jpg")  # Use raw string to handle backslashes

# Display the image
image.show()

```

```python

import gradio as gr

def generate_image(prompt):
    image = pipe(prompt).images[0]
    return image

gr.Interface(
    fn=generate_image,
    inputs="text",
    outputs="image",
    title="Text-to-Image Generator",
    description="Enter a text prompt to generate an image using Stable Diffusion."
).launch()
```

```python
from diffusers import StableDiffusionPipeline
import torch
import gradio as gr

# Load model
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")  # Ensure you're using GPU if available

# Define the Gradio interface
def generate_image(prompt):
    image = pipe(prompt).images[0]
    return image

gr.Interface(
    fn=generate_image,
    inputs="text",
    outputs="image",
    title="Text-to-Image Generator",
    description="Enter a text prompt to generate an image using Stable Diffusion."
).launch()
```
